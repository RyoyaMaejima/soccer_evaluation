```python
# 動画の処理設定
import subprocess
import os
import shutil
from PIL import Image

# 試合データのパス
game_base = "data/SoccerNetData/england_epl/2015-2016/2016-02-07 - 19-00 Chelsea 1 - 1 Manchester United"

# 前半か後半の設定
halfNum = 2

# データのパス
base_path = "data/outputData"

if halfNum == 1:
    game_dir = os.path.join(game_base, f"1_720p.mkv")
    relative_path = os.path.relpath(game_base, "data/SoccerNetData")
    output_base = os.path.join(base_path, relative_path)
    output_base = os.path.join(output_base, f"1")
else:
    game_dir = os.path.join(game_base, f"2_720p.mkv")
    relative_path = os.path.relpath(game_base, "data/SoccerNetData")
    output_base = os.path.join(base_path, relative_path)
    output_base = os.path.join(output_base, f"2")
```

```python
# テキストファイル作成
# ディレクトリが存在しない場合は作成
os.makedirs(output_base, exist_ok=True)

# 作成するファイル名のリスト
filenames = ["AnyNumbers.txt", "DeleteNumbers.txt", "EditNumbers.txt", "FilterNumbers.txt", "Formations.txt"]

# 各ファイルに "fin" を書き込む
for filename in filenames:
    file_path = os.path.join(output_base, filename)
    if not os.path.exists(file_path):
        with open(file_path, "w", encoding="utf-8") as f:
            f.write("fin")
        print(f"{filename} を作成しました")
    else:
        print(f"{filename} は既に存在します")
```

```python
# トリミング
# 入力ファイルと出力ディレクトリの設定
input_file = game_dir
output_dir = os.path.join(output_base, f"trims")
os.makedirs(output_dir, exist_ok=True)

# 動画全体の長さを取得するためのFFmpegコマンド
def get_video_duration(input_file):
    command = [
        'ffprobe',
        '-v', 'error',
        '-show_entries', 'format=duration',
        '-of', 'default=noprint_wrappers=1:nokey=1',
        input_file
    ]
    result = subprocess.run(command, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
    return float(result.stdout.strip())

# 動画全体の長さを取得
duration = get_video_duration(input_file)

# 30秒ごとに分割して保存
segment_duration = 30
current_start = 0
segment_count = 0

while current_start < duration:
    output_file = os.path.join(output_dir, f'trim_{segment_count:03d}.mp4')
    
    # トリミングコマンド
    command = [
        'ffmpeg',
        '-ss', str(current_start),       # 開始時間
        '-i', input_file,               # 入力ファイル
        '-t', str(segment_duration),    # セグメントの長さ
        '-c', 'copy',                   # 再エンコードせずコピー
        output_file
    ]
    
    try:
        subprocess.run(command, check=True)
        print(f'作成されたファイル: {output_file}')
    except subprocess.CalledProcessError as e:
        print(f'エラーが発生しました: {e}')
        break
    
    # 次のセグメントの開始時間
    current_start += segment_duration
    segment_count += 1

print(f'すべてのセグメントが {output_dir} に保存されました！')
```

```python
# フレーム分割
# 入力と出力のディレクトリ設定
input_dir = os.path.join(output_base, f"trims")
output_dir = os.path.join(output_base, f"trim_images")
sample_image_path = "data/outputData/sample/000001.jpg"
os.makedirs(output_dir, exist_ok=True)

# サンプル画像のサイズを取得
with Image.open(sample_image_path) as img:
    width, height = img.size
    
# 1つの動画あたりの画像数（30秒に750枚）
frames_per_video = 750

# 入力ディレクトリ内のすべての動画ファイルを取得
video_files = [f for f in os.listdir(input_dir) if f.endswith(".mp4")]

# 各動画ファイルに対してフレームを抽出
for video_file in video_files:
    video_path = os.path.join(input_dir, video_file)
    
    # 画像出力先ディレクトリの作成
    image_output_dir = os.path.join(output_dir, video_file.split(".")[0])
    image_output_dir = os.path.join(image_output_dir, "img1")
    os.makedirs(image_output_dir, exist_ok=True)

    # 画像の保存処理
    command = [
        "ffmpeg",
        "-i", video_path,                           # 入力ファイル
        "-vf", f"scale={width}:{height}",           # サンプル画像サイズにスケーリング
        "-vframes", str(frames_per_video),          # 1つの動画から750フレームを抽出
        "-q:v", "2",                                # 高品質で保存
        os.path.join(image_output_dir, "%06d.jpg")  # 6桁の連番で保存
        ]

    try:
        subprocess.run(command, check=True)
        print(f"{video_file} から画像を {image_output_dir} に保存しました")
    except subprocess.CalledProcessError as e:
        print(f"エラーが発生しました: {e}")
        
print("すべての画像の処理が完了しました！")
```

```python
# 動画の合成
# 出力ディレクトリとファイルのパス
input_dir = os.path.join(model_base, f"videos")
output_file = os.path.join(output_base, f"tracked_video.mp4")
file_list_path = "file_list.txt"

# 出力リストのファイルを作成
with open(file_list_path, 'w') as file_list:
    # trimsディレクトリ内のすべての動画ファイルをリスト化
    for video_file in sorted(os.listdir(input_dir)):
        if video_file.startswith('trim_') and video_file.endswith('.mp4'):
            file_list.write(f"file '{os.path.join(input_dir, video_file)}'\n")

# FFmpegコマンドで動画を結合
command = [
    'ffmpeg',
    '-f', 'concat',             # concatモード
    '-safe', '0',               # ファイルパスが安全でない場合に必要
    '-i', file_list_path,       # ファイルリスト
    '-c', 'copy',               # 再エンコードせずにコピー
    output_file                 # 出力ファイルのパス
]

try:
    subprocess.run(command, check=True)
    print(f'動画が結合され、{output_file} に保存されました')
except subprocess.CalledProcessError as e:
    print(f'エラーが発生しました: {e}')
finally:
    # 一時的なファイルリストを削除
    os.remove(file_list_path)
```
